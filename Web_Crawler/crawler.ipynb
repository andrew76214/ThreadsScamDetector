```python
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium import webdriver
import random
import time

# 使用 Chrome 瀏覽器
service = webdriver.ChromeService()
driver = webdriver.Chrome(service=service)

driver.get('https://www.threads.net/')
time.sleep(20)
```

```python
def random_wait(min_wait=3, max_wait=6):
    time.sleep(random.randint(min_wait, max_wait))
```

```python
from selenium.webdriver.common.by import By
# xpath= '//*[@id="barcelona-page-layout"]/div/div/div[2]/div[1]/div[3]/div/div[1]/div[1]'
# 量化交易策略

#195
for i in range (1, 601):
    try:
        xpath = f'//*[@id="barcelona-page-layout"]/div/div/div[2]/div[1]/div[1]/div/div[2]/div/div/div[1]/div[{i}]'
        element = driver.find_element(By.XPATH, xpath)
        if element.text == '':
            # driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            driver.execute_script("window.scrollBy(0, 1500);")
            time.sleep(2)
            element = driver.find_element(By.XPATH, xpath)
            while element.text == '':
                print(f'stop now {i}')
                driver.execute_script("window.scrollBy(0, 1500);")
                time.sleep(1)
                element = driver.find_element(By.XPATH, xpath)
        content.append(element.text)
            #break
    except:
        # driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")  # 每次滾動像素
        driver.execute_script("window.scrollBy(0, 2000);")
        random_wait()
        content.append(element.text)
        while element.text == '':
            # driver.execute_script("window.scrollBy(0, 800);")
            random_wait()
            print(f'stop now1 {i}')
            element = driver.find_element(By.XPATH, xpath)
        content.append(element.text)
```
