```python
def store_post(element, url):
    try:
        username = element.find_element(By.XPATH, './/div[2]/div/div/span').text.strip()
    except:
        username = "!!抓取失敗!!"

    try:
        user_verified = element.find_element(By.CSS_SELECTOR, 'svg[aria-label="已驗證"]')
        user_verified = True

    except:
        user_verified = False
    
    try:
        published_on = element.find_element(By.TAG_NAME, "time").get_attribute("title")

    except:
        published_on = "!!抓取失敗!!"
    
    try:
        try:
            replying = element.find_element(By.XPATH, './/div[3]/div/div[1]/span/span')
            text_element = element.find_element(By.XPATH, './/div[3]/div/div[2]')
            text = text_element.text.strip()
            
        except:
            text_element = element.find_element(By.XPATH, './/div[3]/div/div[1]')
            text = text_element.text.strip()

        if text.endswith("翻譯"):
            text = text[:-3].strip()  # 移除最後兩個字

        try:
            tags = []
            tag_elements = text_element.find_elements(By.TAG_NAME, "a")

            for tag_element in tag_elements:
                tag = tag_element.text.strip()
                tags.append(tag)

        except:
            tags = []

    except:
        text = "!!抓取失敗!!"
        tags = []

    
    
    try:
        images = element.find_elements(By.TAG_NAME, "picture")
        img_count = len(images)

        img_infos = []
        img_urls = []
        img_ids = []

        save_dir = f"image/{current_time}"
        os.makedirs(save_dir, exist_ok=True)

        for image in images:
            try:
                img_info = image.find_element(By.TAG_NAME, "img").get_attribute("alt")
                img_infos.append(img_info)
            except:
                img_infos.append("!!抓取失敗!!")
            try:
                img_url = image.find_element(By.TAG_NAME, "img").get_attribute("src")
                img_urls.append(img_url)
                response = requests.get(img_url, stream=True)

                if response.status_code == 200:
                    filename = img_url.split("/")[-1].split("?")[0]
                    file_path = os.path.join(save_dir, filename)
                    
                    with open(file_path, "wb") as file:
                        for chunk in response.iter_content(1024):
                            file.write(chunk)

                    img_ids.append(filename)
                    print(f"圖片已下載: {filename}")
                else:
                    img_ids.append("!!抓取失敗!!")
                    print("下載失敗，狀態碼:", response.status_code)

            except:
                img_urls.append("!!抓取失敗!!")
                img_ids.append("!!抓取失敗!!")
            
    except:
        img_count = 0
        img_infos = []
        img_urls = []
        img_ids = []

    try:
        like_count = element.find_element(By.CSS_SELECTOR, 'svg[aria-label="讚"]').find_element(By.XPATH, "..").text.strip()
        if like_count == '':
            like_count = 0
        else:
            like_count = int(like_count)
    except:
        like_count = "!!抓取失敗!!"
    
    try:
        reply_count = element.find_element(By.CSS_SELECTOR, 'svg[aria-label="回覆"]').find_element(By.XPATH, "..").text.strip()
        if reply_count == '':
            reply_count = 0
        else:
            reply_count = int(reply_count)
    except:
        reply_count = "!!抓取失敗!!"
    
    try:
        forward_count = element.find_element(By.CSS_SELECTOR, 'svg[aria-label="轉發"]').find_element(By.XPATH, "..").text.strip()
        if forward_count == '':
            forward_count = 0
        else:
            forward_count = int(forward_count)
    except:
        forward_count = "!!抓取失敗!!"
    
    try:
        share_count = element.find_element(By.CSS_SELECTOR, 'svg[aria-label="分享"]').find_element(By.XPATH, "..").text.strip()
        if share_count == '':
            share_count = 0
        else:
            share_count = int(share_count)
    except:
        share_count = "!!抓取失敗!!"
        
    
    post = {
        "username": username,
        "user_verified" : user_verified,
        "published_on": published_on,
        "text": text,
        "tags" : tags,
        "img_count" : img_count,
        "img_infos" : img_infos,
        "img_urls" : img_urls,
        "img_ids" : img_ids,
        "like_count": like_count,
        "reply_count": reply_count,
        "forward_count": forward_count,
        "share_count": share_count,
        "url" : url
    }
    
    return post
```

```python
content = []
urls = set()
current_time = datetime.now().strftime("%m%d_%H%M")
target_posts = 50  # 目標貼文數量
max_scrolls = 20  # 最多滾動次數，避免無限迴圈

scroll_count = 0  # 計算滾動次數
while len(content) < target_posts and scroll_count < max_scrolls:
    scroll_count += 1
    print(f"🔄 第 {scroll_count} 次滾動...")

    # **1️⃣ 找到目前頁面所有貼文**
    post_elements = driver.find_elements(By.XPATH, '//*[@id="barcelona-page-layout"]/div/div/div[2]/div[1]/div[1]/div/div[2]/div/div/div[1]/div')

    for post_element in post_elements:
        try:

            try:
                # **2️⃣ 檢查該貼文是否已擷取**
                url = post_element.find_element(By.TAG_NAME, "time").find_element(By.XPATH, "..").get_attribute("href")

                if url in urls:
                    continue

                urls.add(url)

            except:
                continue

            post_data = store_post(post_element, url)
            if post_data and any(post_data.values()) and post_data not in content:
                content.append(post_data)
                print(f"✅ 擷取第 {len(content)} 個貼文: {post_data}")

            if len(content) >= target_posts:
                break  # **提早結束迴圈**
        except Exception as e:
            print(f"⚠️ 擷取失敗: {e}")
            continue  # **跳過此貼文**

    # **3️⃣ 滾動到底部**
    driver.execute_script("window.scrollBy(0, 2000);")
    time.sleep(2)

print(f"🎯 總共找到 {len(content)} 個貼文")
```

```python
output_path = f"output_json/data_{current_time}.json"

with open(output_path, "w", encoding="utf-8") as json_file:
    json.dump(content, json_file, ensure_ascii=False, indent=4)

print(f"✅ JSON 檔案已儲存至: {output_path}")
```
