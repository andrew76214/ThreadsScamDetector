```python
def store_post(element, url):
    try:
        username = element.find_element(By.XPATH, './/div[2]/div/div/span').text.strip()
    except:
        username = "!!æŠ“å–å¤±æ•—!!"

    try:
        user_verified = element.find_element(By.CSS_SELECTOR, 'svg[aria-label="å·²é©—è­‰"]')
        user_verified = True

    except:
        user_verified = False
    
    try:
        published_on = element.find_element(By.TAG_NAME, "time").get_attribute("title")

    except:
        published_on = "!!æŠ“å–å¤±æ•—!!"
    
    try:
        try:
            replying = element.find_element(By.XPATH, './/div[3]/div/div[1]/span/span')
            text_element = element.find_element(By.XPATH, './/div[3]/div/div[2]')
            text = text_element.text.strip()
            
        except:
            text_element = element.find_element(By.XPATH, './/div[3]/div/div[1]')
            text = text_element.text.strip()

        if text.endswith("ç¿»è­¯"):
            text = text[:-3].strip()  # ç§»é™¤æœ€å¾Œå…©å€‹å­—

        try:
            tags = []
            tag_elements = text_element.find_elements(By.TAG_NAME, "a")

            for tag_element in tag_elements:
                tag = tag_element.text.strip()
                tags.append(tag)

        except:
            tags = []

    except:
        text = "!!æŠ“å–å¤±æ•—!!"
        tags = []

    
    
    try:
        images = element.find_elements(By.TAG_NAME, "picture")
        img_count = len(images)

        img_infos = []
        img_urls = []
        img_ids = []

        save_dir = f"image/{current_time}"
        os.makedirs(save_dir, exist_ok=True)

        for image in images:
            try:
                img_info = image.find_element(By.TAG_NAME, "img").get_attribute("alt")
                img_infos.append(img_info)
            except:
                img_infos.append("!!æŠ“å–å¤±æ•—!!")
            try:
                img_url = image.find_element(By.TAG_NAME, "img").get_attribute("src")
                img_urls.append(img_url)
                response = requests.get(img_url, stream=True)

                if response.status_code == 200:
                    filename = img_url.split("/")[-1].split("?")[0]
                    file_path = os.path.join(save_dir, filename)
                    
                    with open(file_path, "wb") as file:
                        for chunk in response.iter_content(1024):
                            file.write(chunk)

                    img_ids.append(filename)
                    print(f"åœ–ç‰‡å·²ä¸‹è¼‰: {filename}")
                else:
                    img_ids.append("!!æŠ“å–å¤±æ•—!!")
                    print("ä¸‹è¼‰å¤±æ•—ï¼Œç‹€æ…‹ç¢¼:", response.status_code)

            except:
                img_urls.append("!!æŠ“å–å¤±æ•—!!")
                img_ids.append("!!æŠ“å–å¤±æ•—!!")
            
    except:
        img_count = 0
        img_infos = []
        img_urls = []
        img_ids = []

    try:
        like_count = element.find_element(By.CSS_SELECTOR, 'svg[aria-label="è®š"]').find_element(By.XPATH, "..").text.strip()
        if like_count == '':
            like_count = 0
        else:
            like_count = int(like_count)
    except:
        like_count = "!!æŠ“å–å¤±æ•—!!"
    
    try:
        reply_count = element.find_element(By.CSS_SELECTOR, 'svg[aria-label="å›è¦†"]').find_element(By.XPATH, "..").text.strip()
        if reply_count == '':
            reply_count = 0
        else:
            reply_count = int(reply_count)
    except:
        reply_count = "!!æŠ“å–å¤±æ•—!!"
    
    try:
        forward_count = element.find_element(By.CSS_SELECTOR, 'svg[aria-label="è½‰ç™¼"]').find_element(By.XPATH, "..").text.strip()
        if forward_count == '':
            forward_count = 0
        else:
            forward_count = int(forward_count)
    except:
        forward_count = "!!æŠ“å–å¤±æ•—!!"
    
    try:
        share_count = element.find_element(By.CSS_SELECTOR, 'svg[aria-label="åˆ†äº«"]').find_element(By.XPATH, "..").text.strip()
        if share_count == '':
            share_count = 0
        else:
            share_count = int(share_count)
    except:
        share_count = "!!æŠ“å–å¤±æ•—!!"
        
    
    post = {
        "username": username,
        "user_verified" : user_verified,
        "published_on": published_on,
        "text": text,
        "tags" : tags,
        "img_count" : img_count,
        "img_infos" : img_infos,
        "img_urls" : img_urls,
        "img_ids" : img_ids,
        "like_count": like_count,
        "reply_count": reply_count,
        "forward_count": forward_count,
        "share_count": share_count,
        "url" : url
    }
    
    return post
```

```python
content = []
urls = set()
current_time = datetime.now().strftime("%m%d_%H%M")
target_posts = 50  # ç›®æ¨™è²¼æ–‡æ•¸é‡
max_scrolls = 20  # æœ€å¤šæ»¾å‹•æ¬¡æ•¸ï¼Œé¿å…ç„¡é™è¿´åœˆ

scroll_count = 0  # è¨ˆç®—æ»¾å‹•æ¬¡æ•¸
while len(content) < target_posts and scroll_count < max_scrolls:
    scroll_count += 1
    print(f"ğŸ”„ ç¬¬ {scroll_count} æ¬¡æ»¾å‹•...")

    # **1ï¸âƒ£ æ‰¾åˆ°ç›®å‰é é¢æ‰€æœ‰è²¼æ–‡**
    post_elements = driver.find_elements(By.XPATH, '//*[@id="barcelona-page-layout"]/div/div/div[2]/div[1]/div[1]/div/div[2]/div/div/div[1]/div')

    for post_element in post_elements:
        try:

            try:
                # **2ï¸âƒ£ æª¢æŸ¥è©²è²¼æ–‡æ˜¯å¦å·²æ“·å–**
                url = post_element.find_element(By.TAG_NAME, "time").find_element(By.XPATH, "..").get_attribute("href")

                if url in urls:
                    continue

                urls.add(url)

            except:
                continue

            post_data = store_post(post_element, url)
            if post_data and any(post_data.values()) and post_data not in content:
                content.append(post_data)
                print(f"âœ… æ“·å–ç¬¬ {len(content)} å€‹è²¼æ–‡: {post_data}")

            if len(content) >= target_posts:
                break  # **ææ—©çµæŸè¿´åœˆ**
        except Exception as e:
            print(f"âš ï¸ æ“·å–å¤±æ•—: {e}")
            continue  # **è·³éæ­¤è²¼æ–‡**

    # **3ï¸âƒ£ æ»¾å‹•åˆ°åº•éƒ¨**
    driver.execute_script("window.scrollBy(0, 2000);")
    time.sleep(2)

print(f"ğŸ¯ ç¸½å…±æ‰¾åˆ° {len(content)} å€‹è²¼æ–‡")
```
